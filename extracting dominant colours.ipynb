```python
##image segmentation partitions the image into regions having similar visual appearance corresponding to parts of object
##we will try to extract the most dominant k colours using the k-means
##We will apply kmeans with each pixel will reassigned to the closest of k colours leading to segmentation
#so in this project we will try to group together the regions having the similour colour
#k=2 means that our image will have only 2 colours and so on so k=10 will mean that we want our image to have 10 colours  
```

```python
import cv2
import numpy as np
```

```python
im= cv2.imread('elephant.jpg') #opencv reads the image into bgr format
print(im.shape)
```

```python
import matplotlib.pyplot as plt 
```

```python
plt.imshow(im) #matplotlib reads the image into Rgb format
plt.show()
```

```python
im=cv2.cvtColor(im,cv2.COLOR_BGR2RGB) #as we have converted the image from bgr to rgb we can see that now it is shown clearly
```

```python
plt.imshow(im) #matplotlib reads the image into Rgb format
originalshape=im.shape
plt.show()
```

```python
all_pixels=im.reshape((-1,3))
print(all_pixels.shape)
```

```python
from sklearn.cluster import KMeans
```

```python
dominant_colours=4
km=KMeans(n_clusters=dominant_colours)
km.fit(all_pixels)
```

```python
centers=km.cluster_centers_
```

```python
print(centers)
```

```python
i=1
plt.figure(0,figsize=(8,2))
colors=[]

for each_col in centers:
    plt.subplot(1,4,i)
    plt.axis("off")
    i+=1
    
    colors.append(each_col)
    a=np.zeros((100,100,3),dtype='uint8')
    a[:,:,:]=each_col
    plt.imshow(a)
plt.show()
#these are the four mostdominant colour in theimage
```

```python
new_img=np.zeros((330*500,3),dtype='uint8')#so using the kmeans we will assign each pixel a new colour out of the above four colours
print(new_img.shape)
```

```python
colors
```

```python
km.labels_
```

```python
for ix in range(new_img.shape[0]):
    new_img[ix]=colors[km.labels_[ix]]
new_img=new_img.reshape((originalshape))
plt.imshow(new_img)
plt.show()
#as we can see that we have created the image only using the above 4 colurs 
```

```python
from scipy.misc import toimage
ex =toimage(new_img)
ex.save("new.jpg")
```
